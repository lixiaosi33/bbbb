```python
import hmac,random,time
import hashlib
import json, requests

def getSign(data,secret):
    result = hmac.new(secret.encode("utf-8"), data.encode("utf-8"), hashlib.md5).hexdigest()
    return result

def doApiRequestWithApikey(url, cmds, api_key, api_secret):
    s_cmds = json.dumps(cmds)
    sign = getSign(s_cmds,api_secret)
    r = requests.post(url, data={'cmds': s_cmds, 'apikey': api_key,'sign':sign})
    print(r.text)

def post_order(api_key,api_secret,cmds):
    url = "https://api.bibox.com/v1/orderpending"
    doApiRequestWithApikey(url,cmds,api_key,api_secret)



api_key = '49d43ef25d6dca7c8396a36fbb368e8255ca2ac1'
api_secret = '4e9b9730a7cf5b3015e90346d65681aa15441959'

def buy(price_buy,vol_buy):
    account_type = 0 
    order_type = 2 
    order_side = 1 
    pair = 'CHS_USDT'
    #price = 
    #amount = 1

    cmds = [
        {
            'cmd':"orderpending/trade",
            'body':{
                'pair':pair,
                'account_type':account_type,
                'order_type':order_type,
                'order_side':order_side,
                'price':price_buy ,
                'amount':vol_buy
            }
        }
    ]
    post_order(api_key,api_secret,cmds)
    
def sell(price_buy,vol_buy):
    account_type = 0 
    order_type = 2 
    order_side = 2 
    pair = 'CHS_USDT'
    #price = 
    #amount = 1

    cmds = [
        {
            'cmd':"orderpending/trade",
            'body':{
                'pair':pair,
                'account_type':account_type,
                'order_type':order_type,
                'order_side':order_side,
                'price':price_buy ,
                'amount':vol_buy
            }
        }
    ]
    post_order(api_key,api_secret,cmds)
    
def get_price():
    url = "https://api.bibox.com/v1/mdata?cmd=depth&pair=CHS_USDT&size=10"
    r = requests.get(url)
    res = r.text
    rese = eval(res)
    ask1 = rese['result']['asks'][0]['price']
    avol1 = rese['result']['asks'][0]['volume']
    bid1 = rese['result']['bids'][0]['price']
    bvol1 = rese['result']['bids'][0]['volume']
    return float(ask1), float(bid1),float(avol1), float(bvol1)



count = 0
price = 0

def combine(pr,st,b1,a1):
    print ("pr before",pr,"st",st,a1,b1)
    bei = random.randint(-2, 2) 
    res = pr+st*bei
    print ("res_before",res,bei,st)
    st1 = 1
    if res>=(a1-st1):
        res = a1-st1
    if res<=(b1+st1):
        res = b1 + st1
    #print ("bei",bei,"res",res)
    print ("res_after",res)
    return int(res)    

while True:
    try:
        slee1 = random.randint(30,50)
        slee2 = random.randint(10,50)
        slee3 = random.randint(40,100)
        slee = random.choice([slee1,slee2,slee3])
        print ("sleep: "+str(slee)+"s")
        time.sleep(slee)
        count = count + 1
        ask1,bid1, avol1, bvol1 = get_price()
        print ("ask1",ask1,"bid1",bid1, "avol1",avol1,"bvol1",bvol1)
        if (avol1>40) & (bvol1>40):
            if count %1==0:
                if ask1*1e4-bid1*1e4>=5:
                    if price ==0:
                        price = int((ask1*1e4+bid1*1e4)/2)*1.0/1e4
                    else:
                        step = random.randint(2,10)*5
                        deltaP = combine(int(float(price)*1e4),step,bid1*1e4,ask1*1e4)
                        price = deltaP/1e4
                    print ("============price================",price)
                    dot = random.randint(1,10)
                    if dot in [1,2,3,4]:
                        vol = random.randint(10,200)
                    elif dot in [5,6,7]:
                        vol = random.randint(10,200) + random.randint(1,10)*1000/10000.0
                    elif dot in [8,9]:
                        vol = random.randint(10,200) + random.randint(1,100)*100/10000.0
                    else:
                        vol = random.randint(10,200) + random.randint(1,1000)*10/10000.0
                    print ("vol=========  ",vol)
                    #print ("make orders, price", price)
                    dir_flag = random.randint(0,1)
                    if dir_flag==0:
                        buy(price,vol)
                        sell(price,vol)    
                        print ("buy-sell:",price,vol)
                    else:
                        sell(price,vol) 
                        buy(price,vol)
                        print ("sell-buy:",price,vol)     
            else:
                pass
    except:
        pass
```
