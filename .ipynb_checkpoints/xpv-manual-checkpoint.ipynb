```python
def cov0(x, y):
    return np.nanmean(np.multiply(x, y)) - np.nanmean(x, axis=0) * np.nanmean(y, axis=0)


def corr(x, y):
    return cov0(x, y) / np.std(x, axis=0) / np.std(y, axis=0)

def cov1(x, y):
    return corr(x, y)* np.std(x, axis=0,ddof=1) * np.std(y, axis=0,ddof=1)



def cor(x,y,n):
    A=np.tile(np.nan,x.shape)
    for i in range(n,len(x)+1):
        x1=x[i-n:i,:]
        y1=y[i-n:i,:]
        x2=(x1-x1.mean(axis=0))/x1.std(axis=0)
        y2=(y1-y1.mean(axis=0))/y1.std(axis=0)
        A[i-1,:]=(x2*y2).mean(axis=0)
    return A

def covar(a,b):
    #计算协方差
    a_disper = a - a.mean(axis=0)
    b_disper = b - b.mean(axis=0)
    cov_ab = (a_disper * b_disper).mean(axis=0)
    
    return cov_ab

def std(A,n):
    return pd.DataFrame(A).rolling(n).std().values

def cov2(x, y, n):
    return cor(x, y,n)[-1]* np.std(x, axis=0,ddof=1) * np.std(y, axis=0,ddof=1)

def cov3(x, y, n):
    return cor(x, y,n) * std(x,n) * std(y,n)

def cov4(x,y,n):
        A=np.tile(np.nan,x.shape)
        for i in range(n,len(x)+1):
                x1=x[i-n:i,:]
                y1=y[i-n:i,:]
                x2=(x1-x1.mean(axis=0))
                y2=(y1-y1.mean(axis=0))
                A[i-1,:]=(x2*y2).mean(axis=0)
        return A*(n)/(n-1)

    
def PearsonCorr(self,a,b):
    #计算Pearson相关性系数
    return self.covar(a,b)/np.sqrt(self.covar(a,a)*self.covar(b,b))
```

```python
def ema(x,n):
    if (n==1):
        return x
    alpha = 1.0/(n+1)
    lastfactor = np.full(x.shape[1], np.nan, dtype=np.float64)
    res = np.full(x.shape, np.nan, dtype=np.float64)
    for i in range(len(x)):
        valid = np.isfinite(x[i,:])
        lastvalid = np.isfinite(lastfactor[valid])
        tem = x[i,:][valid]
        tem[lastvalid] = alpha* tem[lastvalid] + (1-alpha)*(lastfactor[valid])[lastvalid]
        res[i,:][valid] = tem[:]
        res[i,:][~valid] = lastfactor[~valid]
        #print('res', res)
        lastfactor[:] = res[i,:]
    return res
        
```

```python
# def ewmcov(float64_t[:] input_x, float64_t[:] input_y,
#            float64_t com, int adjust, int ignore_na, int minp, int bias):
def ewmcov(input_x, input_y,com, adjust, ignore_na, minp, bias):
    """
    Compute exponentially-weighted moving variance using center-of-mass.
    Parameters
    ----------
    input_x : ndarray (float64 type)
    input_y : ndarray (float64 type)
    com : float64
    adjust: int
    ignore_na: int
    minp: int
    bias: int
    Returns
    -------
    y : ndarray
    """

#     cdef:
#         Py_ssize_t N = len(input_x)
#         float64_t alpha, old_wt_factor, new_wt, mean_x, mean_y, cov
#         float64_t sum_wt, sum_wt2, old_wt, cur_x, cur_y, old_mean_x, old_mean_y
#         Py_ssize_t i, nobs
#         ndarray[float64_t] output

#     if <Py_ssize_t>len(input_y) != N:
#         raise ValueError("arrays are of different lengths "
#                          "({N} and {len_y})".format(N=N, len_y=len(input_y)))
    N = len(input_x)
    output = np.empty(N, dtype=float)
    if N == 0:
        return output

    minp = max(minp, 1)

    alpha = 1. / (1. + com)
    old_wt_factor = 1. - alpha
    new_wt = 1. if adjust else alpha

    mean_x = input_x[0]
    mean_y = input_y[0]
    is_observation = ((mean_x == mean_x) and (mean_y == mean_y))
    
    nobs = int(is_observation)
    if not is_observation:
        mean_x = np.NaN
        mean_y = np.NaN
    output[0] = (0. if bias else np.NaN) if (nobs >= minp) else np.NaN
    cov = 0.
    sum_wt = 1.
    sum_wt2 = 1.
    old_wt = 1.

    for i in range(1, N):
        cur_x = input_x[i]
        cur_y = input_y[i]
        is_observation = ((cur_x == cur_x) and (cur_y == cur_y))
        nobs += int(is_observation)
        if mean_x == mean_x:
            if is_observation or (not ignore_na):
                sum_wt *= old_wt_factor
                sum_wt2 *= (old_wt_factor * old_wt_factor)
                old_wt *= old_wt_factor
                if is_observation:
                    old_mean_x = mean_x
                    old_mean_y = mean_y
                    if mean_x != cur_x:
                        mean_x = ((old_wt * old_mean_x) +
                                  (new_wt * cur_x)) / (old_wt + new_wt)

                    # avoid numerical errors on constant series
                    if mean_y != cur_y:
                        mean_y = ((old_wt * old_mean_y) +
                                  (new_wt * cur_y)) / (old_wt + new_wt)
                    cov = ((old_wt * (cov + ((old_mean_x - mean_x) *
                                             (old_mean_y - mean_y)))) +
                           (new_wt * ((cur_x - mean_x) *
                                      (cur_y - mean_y)))) / (old_wt + new_wt)
                    
                    sum_wt += new_wt
                    sum_wt2 += (new_wt * new_wt)
                    old_wt += new_wt
                    if not adjust:
                        sum_wt /= old_wt
                        sum_wt2 /= (old_wt * old_wt)
                        old_wt = 1.
        elif is_observation:
            mean_x = cur_x
            mean_y = cur_y

        if nobs >= minp:
            if not bias:
                numerator = sum_wt * sum_wt
                denominator = numerator - sum_wt2
                if (denominator > 0.):
                    output[i] = ((numerator / denominator) * cov)
                else:
                    output[i] = np.NaN
            else:
                output[i] = cov
        else:
            output[i] = np.NaN
    return output
```

```python
def ewmcovN(aa,bb,com):
    res = np.full(aa.shape, np.nan, dtype=np.float64)
    for ii in range(aa.shape[1]):
        aaN = aa[:,ii]
        bbN = bb[:,ii]
        res[:,ii] = ewmcov(aaN, bbN,com=com, adjust=True, ignore_na=False, minp=0, bias=False)
    return res

def ewmstdN(aa,com):
    res = np.full(aa.shape, np.nan, dtype=np.float64)
    for ii in range(aa.shape[1]):
        aaN = aa[:,ii]
        res[:,ii] = np.sqrt(ewmcov(aaN, aaN,com=com, adjust=True, ignore_na=False, minp=0, bias=False))
    return res

def ewmcorrN(aa,bb,com):
    res = np.full(aa.shape, np.nan, dtype=np.float64)
    rescov = np.full(aa.shape, np.nan, dtype=np.float64)
    resstd1 = np.full(aa.shape, np.nan, dtype=np.float64)
    resstd2 = np.full(aa.shape, np.nan, dtype=np.float64)
    for ii in range(aa.shape[1]):
        aaN = aa[:,ii]
        bbN = bb[:,ii]
        rescov[:,ii] = ewmcov(aaN, bbN,com=com, adjust=True, ignore_na=False, minp=0, bias=False)
        resstd1[:,ii] = np.sqrt(ewmcov(aaN, aaN,com=com, adjust=True, ignore_na=False, minp=0, bias=False))
        resstd2[:,ii] = np.sqrt(ewmcov(bbN, bbN,com=com, adjust=True, ignore_na=False, minp=0, bias=False))
    res = rescov/(resstd1 * resstd2)
    return res
```

```python
# def ewmcov(float64_t[:] input_x, float64_t[:] input_y,
#            float64_t com, int adjust, int ignore_na, int minp, int bias):
def cur_ewmcov(cur_x, cur_y,com,sum_wt,sum_wt2,old_wt,mean_x,mean_y,cov,  adjust=True, ignore_na=False, minp=0, bias=False ):
    """
    Compute exponentially-weighted moving variance using center-of-mass.
    Parameters
    ----------
    input_x : ndarray (float64 type)
    input_y : ndarray (float64 type)
    com : float64
    adjust: int
    ignore_na: int
    minp: int
    bias: int
    Returns
    -------
    y : ndarray
    """

#     cdef:
#         Py_ssize_t N = len(input_x)
#         float64_t alpha, old_wt_factor, new_wt, mean_x, mean_y, cov
#         float64_t sum_wt, sum_wt2, old_wt, cur_x, cur_y, old_mean_x, old_mean_y
#         Py_ssize_t i, nobs
#         ndarray[float64_t] output

#     if <Py_ssize_t>len(input_y) != N:
#         raise ValueError("arrays are of different lengths "
#                          "({N} and {len_y})".format(N=N, len_y=len(input_y)))



    minp = max(minp, 1)

    alpha = 1. / (1. + com)
    old_wt_factor = 1. - alpha


    new_wt = 1. if adjust else alpha
#     is_observation = ((mean_x == mean_x) and (mean_y == mean_y))
    
#     nobs = int(is_observation)
#     if not is_observation:
#         mean_x = np.NaN
#         mean_y = np.NaN
#     print ("before","is_observation",is_observation, "nobs",nobs)

#===========

    is_observation = ((cur_x == cur_x) and (cur_y == cur_y))
    if mean_x == mean_x:
        if is_observation or (not ignore_na):
            sum_wt *= old_wt_factor
            sum_wt2 *= (old_wt_factor * old_wt_factor)
            old_wt *= old_wt_factor
            if is_observation:
                old_mean_x = mean_x
                old_mean_y = mean_y

                if mean_x != cur_x:
                    mean_x = ((old_wt * old_mean_x) +
                              (new_wt * cur_x)) / (old_wt + new_wt)

                # avoid numerical errors on constant series
                if mean_y != cur_y:
                    mean_y = ((old_wt * old_mean_y) +
                              (new_wt * cur_y)) / (old_wt + new_wt)
                cov = ((old_wt * (cov + ((old_mean_x - mean_x) *
                                         (old_mean_y - mean_y)))) +
                       (new_wt * ((cur_x - mean_x) *
                                  (cur_y - mean_y)))) / (old_wt + new_wt)
                sum_wt += new_wt
                sum_wt2 += (new_wt * new_wt)
                old_wt += new_wt
                if not adjust:
                    sum_wt /= old_wt
                    sum_wt2 /= (old_wt * old_wt)
                    old_wt = 1.
    elif is_observation:
        mean_x = cur_x
        mean_y = cur_y
        
    nobs = 1
    if nobs >= minp:
        if not bias:
            numerator = sum_wt * sum_wt
            denominator = numerator - sum_wt2
            if (denominator > 0.):
                output = ((numerator / denominator) * cov)
            else:
                output = np.NaN
        else:
            output = cov
    else:
        output = np.NaN
    return sum_wt,sum_wt2,old_wt,mean_x,mean_y,cov,output
```

```python
def ewmcovN(aa,bb,com,mean_x_list,mean_y_list,cov_list,sum_wt_list,sum_wt2_list,old_wt_list):
    res = np.full(aa.shape, np.nan, dtype=np.float64)
    for ii in range(len(aa)):
        cur_x,cur_y = aa[ii],bb[ii]
        sum_wt,sum_wt,sum_wt2,old_wt = sum_wt_list[ii],sum_wt_list[ii],sum_wt2_list[ii],old_wt_list[ii]
        mean_x,mean_y,cov = mean_x_list[ii],mean_y_list[ii],cov_list[ii]
        sum_wt,sum_wt2,old_wt,mean_x,mean_y,cov,output = cur_ewmcov(cur_x, cur_y,com,sum_wt,sum_wt2,old_wt,mean_x,mean_y,cov,  adjust=True, ignore_na=False, minp=0, bias=False )
        sum_wt_list[ii],sum_wt_list[ii],sum_wt2_list[ii],old_wt_list[ii] = sum_wt,sum_wt,sum_wt2,old_wt
        mean_x_list[ii],mean_y_list[ii],cov_list[ii] = mean_x,mean_y,cov
        res[ii] = output
    return mean_x_list,mean_y_list,cov_list,sum_wt_list,sum_wt2_list,old_wt_list,res

def ewmstdN(aa,bb,com,std_mean_x_list,std_mean_y_list,std_cov_list,std_sum_wt_list,std_sum_wt2_list,std_old_wt_list):
    res = np.full(aa.shape, np.nan, dtype=np.float64)
    for ii in range(len(aa)):
        cur_x,cur_y = aa[ii],aa[ii]
        std_sum_wt,std_sum_wt,std_sum_wt2,std_old_wt = std_sum_wt_list[ii],std_sum_wt_list[ii],std_sum_wt2_list[ii],std_old_wt_list[ii]
        std_mean_x,std_mean_y,std_cov = std_mean_x_list[ii],std_mean_y_list[ii],std_cov_list[ii]
        std_sum_wt,std_sum_wt2,std_old_wt,std_mean_x,std_mean_y,std_cov,output = cur_ewmcov(cur_x, cur_y,com,std_sum_wt,std_sum_wt2,std_old_wt,std_mean_x,std_mean_y,std_cov,  adjust=True, ignore_na=False, minp=0, bias=False )
        std_sum_wt_list[ii],std_sum_wt_list[ii],std_sum_wt2_list[ii],std_old_wt_list[ii] = std_sum_wt,std_sum_wt,std_sum_wt2,std_old_wt
        std_mean_x_list[ii],std_mean_y_list[ii],std_cov_list[ii] = std_mean_x,std_mean_y,std_cov
        res[ii] = np.sqrt(output)
    return std_mean_x_list,std_mean_y_list,std_cov_list,std_sum_wt_list,std_sum_wt2_list,std_old_wt_list,res

def ewmcorrN(aa,bb,com,corr_mean_x_list,corr_mean_y_list,corr_cov_list,corr_sum_wt_list,corr_sum_wt2_list,corr_old_wt_list\
            ,corr_std1_mean_x_list,corr_std1_mean_y_list,corr_std1_cov_list,corr_std1_sum_wt_list,corr_std1_sum_wt2_list,corr_std1_old_wt_list\
            ,corr_std2_mean_x_list,corr_std2_mean_y_list,corr_std2_cov_list,corr_std2_sum_wt_list,corr_std2_sum_wt2_list,corr_std2_old_wt_list):
    res = np.full(aa.shape, np.nan, dtype=np.float64)
    rescov = np.full(aa.shape, np.nan, dtype=np.float64)
    resstd1 = np.full(aa.shape, np.nan, dtype=np.float64)
    resstd2 = np.full(aa.shape, np.nan, dtype=np.float64)
    for ii in range(len(aa)):
        cur_x,cur_y = aa[ii],bb[ii]
        corr_sum_wt,corr_sum_wt,corr_sum_wt2,corr_old_wt = corr_sum_wt_list[ii],corr_sum_wt_list[ii],corr_sum_wt2_list[ii],corr_old_wt_list[ii]
        corr_mean_x,corr_mean_y,corr_cov = corr_mean_x_list[ii],corr_mean_y_list[ii],corr_cov_list[ii]
        corr_sum_wt,corr_sum_wt2,corr_old_wt,corr_mean_x,corr_mean_y,corr_cov,output1 = cur_ewmcov(cur_x, cur_y,com,corr_sum_wt,corr_sum_wt2,corr_old_wt,corr_mean_x,corr_mean_y,corr_cov,  adjust=True, ignore_na=False, minp=0, bias=False )
        corr_sum_wt_list[ii],corr_sum_wt_list[ii],corr_sum_wt2_list[ii],corr_old_wt_list[ii] = corr_sum_wt,corr_sum_wt,corr_sum_wt2,corr_old_wt
        corr_mean_x_list[ii],corr_mean_y_list[ii],corr_cov_list[ii] = corr_mean_x,corr_mean_y,corr_cov
        rescov[ii] = output1

        corr_std1_sum_wt,corr_std1_sum_wt,corr_std1_sum_wt2,corr_std1_old_wt = corr_std1_sum_wt_list[ii],corr_std1_sum_wt_list[ii],corr_std1_sum_wt2_list[ii],corr_std1_old_wt_list[ii]
        corr_std1_mean_x,corr_std1_mean_y,corr_std1_cov = corr_std1_mean_x_list[ii],corr_std1_mean_y_list[ii],corr_std1_cov_list[ii]
        corr_std1_sum_wt,corr_std1_sum_wt2,corr_std1_old_wt,corr_std1_mean_x,corr_std1_mean_y,corr_std1_cov,output2 = cur_ewmcov(cur_x, cur_x,com,corr_std1_sum_wt,corr_std1_sum_wt2,corr_std1_old_wt,corr_std1_mean_x,corr_std1_mean_y,corr_std1_cov,  adjust=True, ignore_na=False, minp=0, bias=False )
        corr_std1_sum_wt_list[ii],corr_std1_sum_wt_list[ii],corr_std1_sum_wt2_list[ii],corr_std1_old_wt_list[ii] = corr_std1_sum_wt,corr_std1_sum_wt,corr_std1_sum_wt2,corr_std1_old_wt
        corr_std1_mean_x_list[ii],corr_std1_mean_y_list[ii],corr_std1_cov_list[ii] = corr_std1_mean_x,corr_std1_mean_y,corr_std1_cov
        resstd1[ii] = np.sqrt(output2)
        
        corr_std2_sum_wt,corr_std2_sum_wt,corr_std2_sum_wt2,corr_std2_old_wt = corr_std2_sum_wt_list[ii],corr_std2_sum_wt_list[ii],corr_std2_sum_wt2_list[ii],corr_std2_old_wt_list[ii]
        corr_std2_mean_x,corr_std2_mean_y,corr_std2_cov = corr_std2_mean_x_list[ii],corr_std2_mean_y_list[ii],corr_std2_cov_list[ii]
        corr_std2_sum_wt,corr_std2_sum_wt2,corr_std2_old_wt,corr_std2_mean_x,corr_std2_mean_y,corr_std2_cov,output3 = cur_ewmcov(cur_y, cur_y,com,corr_std2_sum_wt,corr_std2_sum_wt2,corr_std2_old_wt,corr_std2_mean_x,corr_std2_mean_y,corr_std2_cov,  adjust=True, ignore_na=False, minp=0, bias=False )
        corr_std2_sum_wt_list[ii],corr_std2_sum_wt_list[ii],corr_std2_sum_wt2_list[ii],corr_std2_old_wt_list[ii] = corr_std2_sum_wt,corr_std2_sum_wt,corr_std2_sum_wt2,corr_std2_old_wt
        corr_std2_mean_x_list[ii],corr_std2_mean_y_list[ii],corr_std2_cov_list[ii] = corr_std2_mean_x,corr_std2_mean_y,corr_std2_cov
        resstd2[ii] = np.sqrt(output3)
        #res[ii] = rescov[ii]#/(resstd1[ii] * resstd2[ii])
        res[ii] = resstd1[ii] * resstd1[ii]
    return corr_mean_x_list,corr_mean_y_list,corr_cov_list,corr_sum_wt_list,corr_sum_wt2_list,corr_old_wt_list\
            ,corr_std1_mean_x_list,corr_std1_mean_y_list,corr_std1_cov_list,corr_std1_sum_wt_list,corr_std1_sum_wt2_list,corr_std1_old_wt_list\
            ,corr_std2_mean_x_list,corr_std2_mean_y_list,corr_std2_cov_list,corr_std2_sum_wt_list,corr_std2_sum_wt2_list,corr_std2_old_wt_list,res
```
